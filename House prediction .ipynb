```python
# Adding target variable to the DataFrame
data['PRICE'] = boston.target
```

```python
# Splitting target variables and independent variables
y = data['PRICE']
X = data.drop(['PRICE'], axis = 1)
```

```python
# Splitting Data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python

from sklearn.linear_model import LinearRegression

lm = LinearRegression()

lm.fit(X_train, y_train)
```

```python
lm.intercept_
```

```python
# Converting the coeffecient values to DataFrame

coeff = pd.DataFrame([X_train.columns, lm.coef_]).T

coeff = coeff.rename(columns = {0:'Attribute', 1:'Coeffecients'})

coeff
```

```python
# Model Prediction on train data

y_pred = lm.predict(X_train)
```

```python
# Model Evaluation train data

print('R^2:', metrics.r2_score(y_train,y_pred))
print('MSE:', metrics.mean_squared_error(y_train, y_pred))
print('MAE:', metrics.mean_absolute_error(y_train, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_train, y_pred)))
```

```python
# Predicting test data with the model

y_test_pred = lm.predict(X_test)
```

```python
# Model Evaluation

print("R^2:", metrics.r2_score(y_test, y_test_pred))
print('MAE:',metrics.mean_absolute_error(y_test, y_test_pred))
print('MSE:',metrics.mean_squared_error(y_test, y_test_pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))
```
