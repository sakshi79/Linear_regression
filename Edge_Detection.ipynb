```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import cv2
import numpy as np

%matplotlib inline
```

```python
image = mpimg.imread('images/curved_lane.jpg')
plt.imshow(image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image, cmap='gray')
```

```python
# Applying a Sobel filter to the image
sobel_x = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])
sobel_y = np.array([[-1, -2, -1],
                   [0, 0, 0],
                   [1, 2, 1]])

filtered_image_y = cv2.filter2D(gray_image, -1, sobel_y)  # -1 means the output image will have same type as input image
plt.imshow(filtered_image_y, cmap='gray')
```

```python
filtered_image_x = cv2.filter2D(gray_image, -1, sobel_x)  
plt.imshow(filtered_image_x, cmap='gray')
```

```python
# Applying both filters
filtered_image = cv2.filter2D(filtered_image_y, -1, sobel_x)  
plt.imshow(filtered_image, cmap='gray')
```

```python
# Create a threshold that sets all the filtered pixels to white
retval, binary_image = cv2.threshold(filtered_image, 50, 255, cv2.THRESH_BINARY)
plt.imshow(binary_image, cmap='gray')
```

```python
image = mpimg.imread('images/city_hall.jpg')
plt.imshow(image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
filtered_image_x = cv2.filter2D(gray_image, -1, sobel_x)
filtered_image = cv2.filter2D(filtered_image_x, -1, sobel_y)  
retval, binary_image = cv2.threshold(filtered_image, 100, 255, cv2.THRESH_BINARY)
plt.imshow(binary_image, cmap='gray')
```

```python
filter = np.array([[-1, 0, 1],
                 [0, 0, 0],
                 [1, 0, -1]])
filtered_image = cv2.filter2D(gray_image, -1, filter)
retval, binary_image = cv2.threshold(filtered_image, 50, 255, cv2.THRESH_BINARY)
plt.imshow(binary_image, cmap='gray')
```
